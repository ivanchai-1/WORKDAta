```python
n = 1000
result = []
for i in range(n):
    a = np.random.binomial(1,0.03,size = 1000).mean()
    b = np.random.binomial(1,0.05,size = 1000).mean()
    result.append((a,b))
```

```python
df = pd.DataFrame(result,columns= ['a','b'])
```

```python
(df.a != df.b).mean()
```
